## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
purchase_data.head()
```

```python
#Get the unique Items 
unique_items=purchase_data["Item ID"].value_counts()
unique_items=unique_items.count()
unique_items
```

```python
#Get Average Price 
Average_price=purchase_data["Price"].mean()

Average_price
```

```python
#Get the number of purchases 
Purchase_number=purchase_data["Purchase ID"].count()
```

```python
#Get the total Revenue
Total_revenue=purchase_data["Price"].sum()
```

```python
#Build the summary table
summary_table=pd.DataFrame({
    "Number of unique items":[unique_items],
    "Average Price":[Average_price],
    "Number of pruchases":[Purchase_number],
    "Total Revenue": [Total_revenue]
})
#Apply format 
summary_table=summary_table.style.format({"Average Price" : "${:20,.2f}","Total Revenue":"${:20,.2f}"})
summary_table
```

```python
#Group values by gender
purchase_data_grouped=purchase_data.groupby(['Gender'])
purchase_data_grouped.count().head()
```

```python
#Extract Total purchase per Person
gender_mean_price = pd.DataFrame(purchase_data_grouped["Price"].sum())
gender_mean_price
```

```python
#Group the values by gender and person 
total_purchase_df = purchase_data.groupby(['Gender','SN'])
#Sum the values of purchase 
total_purchase_df = pd.DataFrame(total_purchase_df["Price"].sum())
#Regroup values by gender 
total_purchase_df = total_purchase_df .groupby(['Gender'])
#Extract mean by values 
total_purchase_df = pd.DataFrame(total_purchase_df["Price"].mean())
total_purchase_df
```

```python
#Join all the data 
merge_df=pd.merge(gender_sum, gender_price_mean, on = "Gender", how = "outer")
merge_df=pd.merge(merge_df,gender_mean_price, on ="Gender", how = "outer")
merge_df=pd.merge(merge_df,total_purchase_df, on ="Gender", how = "outer")
merge_df
```

```python
#Give format and rename cells
merge_df=merge_df.rename(columns={"Purchase ID":"Purchase Count","Price_x":"Average Purchase Price","Price_y":"Total Purchase Value","Price":"Avg Total Purchase per Person"})
merge_df=merge_df.style.format({"Average Purchase Price":"${:20,.2f}","Total Purchase Value":"${:20,.2f}","Avg Total Purchase per Person":"${:20,.2f}"})
merge_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Call the data 
top_spenders=purchase_data
top_spenders.head()
```

```python
#Call only the important columns 
top_spenders=top_spenders[["SN","Price"]]
top_spenders
```

```python
#group by sn
purchase_count= top_spenders.groupby(['SN'])
#Count the number of products per each user
purchase_count = pd.DataFrame(purchase_count["Price"].count())
#Rename Columns 
purchase_count = purchase_count.rename(columns={"Price":"Purchase Count"})
purchase_count
```

```python
#group by sn
average_purchase_price= top_spenders.groupby(['SN'])
#Count the number of products per each user
average_purchase_price = pd.DataFrame(average_purchase_price["Price"].mean())
#Rename Columns 
average_purchase_price = average_purchase_price.rename(columns={"Price":"Average Purchase Price"})

```

```python
#group by sn
total_purchase_value= top_spenders.groupby(['SN'])
#sum the number of products per each user
total_purchase_value = pd.DataFrame(total_purchase_value["Price"].sum())
#Rename Columns 
total_purchase_value = total_purchase_value.rename(columns={"Price":"Total Purchase Value"})



```

```python
#Merge the data to have the summary table 

merge_popular_users=pd.merge(purchase_count, average_purchase_price, on = "SN", how = "outer")
merge_popular_users=pd.merge(merge_popular_users,total_purchase_value, on = "SN", how ="outer")

#order the values

merge_popular_users = merge_popular_users.sort_values("Total Purchase Value", ascending= False)


#format the table

merge_popular_users.head(5).style.format({"Average Purchase Price":"${:20,.2f}","Total Purchase Value":"${:20,.2f}"})


```
